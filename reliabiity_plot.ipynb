```python
import pandas as pd
import matplotlib.pyplot as plt
import random
plt.rcParams['font.size'] = '16'
```

```python
N_classes = 3
N_nodes = 10
```

```python
class_marker = ['o','x','+']
```

```python
x = [random.rand(1, 100) for n in range(N_nodes)]
y = [random.rand(1, 100) for n in range(N_nodes)]
reliability = [random.randint(1, 100) for n in range(N_nodes)]

plt.scatter(x, y)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y)

for i, txt in enumerate(reliability):
    ax.annotate(txt, (x[i]+1, y[i]+1), fontsize=15)

ax.set_ylim(0,105)
ax.set_xlim(0,100)

plt.savefig("example.pdf",bbox_inches="tight")
```

```python
requester_id = random.randint(0,N_nodes-1)
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.scatter(x[requester_id], y[requester_id], color="tab:red", label="Requester")

sum_rel = sum(reliability)-reliability[requester_id]
for i, txt in enumerate(reliability):
    if i!=requester_id:

        ax.annotate(round(txt/sum_rel,3), (x[i]+1, y[i]+1), fontsize=15)


ax.set_ylim(0,105)
ax.set_xlim(0,100)
ax.legend()
plt.savefig("example_1.pdf",bbox_inches="tight")
```
